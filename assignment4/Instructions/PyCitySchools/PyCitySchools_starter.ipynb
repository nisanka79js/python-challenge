```python
school_data.head(10)
```

```python
school_type = school_data.set_index(['school_name'])['type']
school_type
```

```python
# Calulate total students at each school
total_students = school_data_complete['school_name'].value_counts()
total_students
```

```python
# Budget for each school
Each_school_budget = school_data.groupby(['school_name'])['budget'].sum()
Each_school_budget
```

```python
#per student budget at each school
per_student_budget = Each_school_budget / total_students
per_student_budget
```

```python
# Get the average math score for each school
Avg_math = school_data_complete.groupby(['school_name'])['math_score'].mean()
Avg_math = round(Avg_math,2)
Avg_math
```

```python
# Average reading score at each school
Avg_read = school_data_complete.groupby(['school_name'])['reading_score'].mean()
Avg_read = round(Avg_read, 2)
Avg_read
```

```python
# Percentage of students 70% or more in math for each school
# Grades 70 or higher 
mathpass_school = school_data_complete[(school_data_complete['math_score'] >= 70)]
mathpass_school

# group the scores by school and calculate percentage
school_mathpass_percentage = mathpass_school.groupby(["school_name"])["student_name"].count() / total_students * 100
school_mathpass_percentage = round(school_mathpass_percentage, 2)
school_mathpass_percentage
```

```python
# Percentage of students 70% or more in reading for each school
# Grades 70 or higher 
readpass_school = school_data_complete[(school_data_complete['reading_score'] >= 70)]

# group the scores by school and calculate percentage
school_readpass_percentage = readpass_school.groupby(["school_name"])["student_name"].count() / total_students * 100
school_readpass_percentage = round(school_readpass_percentage, 2)
school_readpass_percentage
```

```python
# Overall passing rate for each school
school_overall_pass = school_mathpass_percentage + school_readpass_percentage / 2
school_overall_pass = round(school_overall_pass, 2)
school_overall_pass
```

```python
# Overall passing rate for each school
Overall_pass = school_mathpass_percentage + school_readpass_percentage / 2
Overall_pass = round(school_overall_pass, 2)
Overall_pass
```

```python
# Create a dataframe showing scool summary info

school_results = pd.DataFrame({
    "School Type": school_type,
    "Students": total_students,
    "School Budget": Each_school_budget,
    "Budget per Student": per_student_budget,
    "Average Math Score": Avg_math,
    "Average Reading Score": Avg_read,
    "% Passing Math": school_mathpass_percentage,
    "% Passing Reading": school_readpass_percentage,
    "Overall Pass Rate": Overall_pass
})

school_results

school_results = school_results[["School Type", "Students", "School Budget", "Budget per Student", "Average Math Score",
                                "Average Reading Score", "% Passing Math", "% Passing Reading",
                                "Overall Pass Rate"]]

school_results
```
